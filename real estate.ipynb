```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.linear_model import LogisticRegression
```

```python
import pandas as pd
df = pd.read_csv('Riyadh_Real_estate.csv')
df
```

```python
df.drop(['Unnamed: 0'], axis=1, inplace = True)
```

```python
df.info()
```

```python
df.describe()
```

```python
df.columns
```

```python
df[df.duplicated()]
```

```python
df.drop_duplicates(keep=False,inplace=True)
```

```python
df[df.duplicated()]
```

```python
df['lounges'] = pd.to_numeric(df['lounges'], errors='coerce')
df['lounges'] = df['lounges'].astype(float)
df['lounges'] = df['lounges'].fillna(df['lounges'].median())
```

```python
df['streetWidth'] = pd.to_numeric(df['streetWidth'], errors='coerce')
df['streetWidth'] = df['streetWidth'].astype(float)
df['streetWidth'] = df['streetWidth'].fillna(df['streetWidth'].median())
```

```python
df['price'] = pd.to_numeric(df['price'], errors='coerce')
df['price'] = df['price'].astype(float)
df['price'] = df['price'].fillna(df['price'].median())
```

```python
df['square price'] = pd.to_numeric(df['square price'], errors='coerce')
df['square price'] = df['square price'].astype(float)
df['square price'] = df['square price'].fillna(df['square price'].median())
```

```python
df['apartments'] = pd.to_numeric(df['apartments'], errors='coerce')
df['apartments'] = df['apartments'].astype(float)
df['apartments'] = df['apartments'].fillna(df['apartments'].median())
df.info()
```

```python
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
var_mod = df.select_dtypes(include='object').columns
for i in var_mod:
   df[i] = le.fit_transform(df[i]) 
df
```

```python
df.info()
```

```python
x=df.drop(['square price'],axis=1)
y=df['square price']

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.20,random_state=10)
```

```python
LR=LinearRegression()
#Start Training
LR.fit(x_train,y_train)
```

```python
pre = LR.predict(x_test)
pre
```

```python
import numpy as np
Answer = np.array(y_test)
Answer
```

```python
ans = LR.score(x_test,y_test) 
ans * 100
```
